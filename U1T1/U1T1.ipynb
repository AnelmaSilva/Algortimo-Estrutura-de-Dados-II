**Exercício 2.2**

60% of the emails hitting my inbox is spam. You design a phenomenal spam filter which is able to tell me, with 98% accuracy, whether an email is spam or not: if an email is not spam, the system has a 98% probability of saying so. The filter knows 60% of emails are spam and so it will flag 60% of my emails. Suppose that, at the end of the week, I look in my spam box and see 963 emails. Use Bayes' Theorem to calculate how many of those 963 emails in my spam box I should suspect to be non-spam.

**Resolução 2.2**

Analisando os dados:
- Probabilidade do email ser spam (P(S)) = 60% ou 0.6
- Probabilidade do email não ser spam (P(n_S)) = 40% ou 0.4
- Probabilidade de prever spam (P(PS))  = 98% ou 0.98
- Probabilidade de prever spam sem ser spam (P(Pn_S)) = 2% ou 0.02

Queremos encontrar a probabilidade de um email não ser spam, dado que o filtro previu que ele é spam (P(n_SP)). Para isso usamos o teorema de Bayes que nos diz o seguinte:
- P(n_SP) = (P(Pn_S) * P(n_S)) / P(P) .
Onde P(P) é dado por : P(P) = P(PS) * P(S) + P(Pn_S) * P(n_S).

Substituindo os valores em P(P), temos que:
P(P) = (0.98*0.6) + (0.0.2*0.4) =  0.592.

Agora vamos calcular P(n_SP) :
P(n_SP) = (0.02 * 0.4) / 0.592 = 0.0135 .

Portanto, a probabilidade dos emails marcados como spam não serem spam neste caso é de 1.35%. Este porcentagem representa um total de aproximadamente 13 email, se aplicamos a quantidade total de 963 email

**Resolução 2.3**

O Processo de Markov e a Cifra de César
Um processo de Markov é um sistema em que o estado futuro depende apenas do estado atual e não de estados anteriores. Na cifra de César, cada letra é substituída por outra que está um número fixo de posições à frente no alfabeto. Ao aplicar a cifra de César repetidamente, estamos criando uma cadeia de Markov de primeira ordem.

A cifra de César desloca cada letra do alfabeto por um número fixo de posições. No seu caso, o deslocamento é de 7 posições. Aplicar essa cifra três vezes equivale a um deslocamento total de 21 posições (7 * 3).

Como o alfabeto possui 26 letras, um deslocamento de 21 posições para frente é o mesmo que um deslocamento de 5 posições para trás (26 - 21 = 5).

Função de Massa de Probabilidade (fmp):

A fmp é uma função que associa a cada valor possível de uma variável aleatória discreta a sua probabilidade. Graficamente, ela é representada por barras, onde a altura de cada barra corresponde à probabilidade do valor associado.

Distribuição Acumulada de Probabilidade (dap):

A dap, por sua vez, representa a probabilidade de que a variável aleatória seja menor ou igual a um determinado valor. Ela é uma função não decrescente que varia entre 0 e 1. Graficamente, é representada por uma linha que "sobe" à medida que os valores da variável aumentam.

**Exercício 4.3**

**This network** is a hypergraph, with a hyperedge per line. Transform it in a unipartite network in which each hyperedge is split in edges connecting all nodes in the hyperedge. Then transform it into a bipartite network in which each hyperedge is a node of one type and its nodes connect to it.



**Resolução 4.3**

Antes de resolvermos o problema, vamos definir um hipergrafo, uma rede unipartida e bipartida. Um hipergrafo é uma generalização de um grafo onde uma aresta pode conectar mais de dois vértices.

Uma rede unipartida cada aresta conecta exatamente dois nós. Para transformar o hipergrafo em uma rede unipartida, devemos dividir cada hiperaresta em várias arestas, conectando todos os pares de nós da hiperaresta original. Já a bipartida  possui dois conjuntos de nós, e as arestas conectam nós de um conjunto ao outro. Para transformar o hipergrafo em uma rede bipartida, criaremos um novo conjunto de nós, onde cada nó representa uma hiperaresta original. Então, conectaremos cada nó da hiperaresta original ao novo nó que representa essa hiperaresta. Agora sim, vamos converter o hipergrafo em cada uma das redes citadas.

Convertente para Rede Unipartida:

Para cada hiperaresta, geramos todas as possíveis combinações de pares de nós e as adicionamos como arestas à nova rede. Por exemplo, a primeira hiperaresta "1 2 3 4 5 6" gerará as seguintes arestas na rede unipartida: (1,2), (1,3), (1,4), (1,5), (1,6), (2,3), (2,4), ..., (5,6).

Convertente para Rede Bipartida:

Criamos um novo conjunto de nós, numerados de forma sequencial. Cada novo nó representará uma hiperaresta. Então, para cada hiperaresta, conectamos o novo nó correspondente a todos os nós originais da hiperaresta. Por exemplo, para a hiperaresta "1 2 3 4 5 6", criamos um novo nó (vamos chamá-lo de "A") e conectamos "A" a 1, 2, 3, 4, 5 e 6.
